```python
# turns each column into a list

nytickerlist = list(nyse_nan.ticker)
nyexchangelist = list(nyse_nan.exchange)
nynamelist = list(nyse_nan.name)
nysectorlist = list(nyse_nan.sector)
nyindustrylist = list(nyse_nan.industry)

nastickerlist = list(nasdaq_nan.ticker)
nasexchangelist = list(nasdaq_nan.exchange)
nasnamelist = list(nasdaq_nan.name)
nassectorlist = list(nasdaq_nan.sector)
nasindustrylist = list(nasdaq_nan.industry)
```

```python
def parse_one(single):
    item = single.replace("&#39;", "'")
    item1 = item.replace("'", '')
    item2 = item1.replace(',', '')
    item3 = item2.replace('.', '')
    item4 = item3.replace('-', '')
    item5 = item4.replace('(', '')
    item6 = item5.replace(')', '')
    item7 = item6.replace('/', ' ')
    item8 = item7.replace('&', 'and')
    item9 = item8.replace(':', '')
    item10 = item9.replace('^', '')
    item11 = item10.title()
    return item11
```

```python
# tests parsing function on one item
parse_one(nynamelist[3])
```

```python
# passes each item in each list through the parsing function
# does it for each exchange dataframe


#1
nynames = [] 
i = 0
for thing in nynamelist:
    new = parse_one(nynamelist[i])
    nynames.append(new)
    i += 1
    
#2
nysector = []
j = 0
for thing in nysectorlist:
    new = parse_one(nysectorlist[j])
    nysector.append(new)
    j += 1
    
#3
nyindustry = []
k = 0
for thing in nyindustrylist:
    new = parse_one(nyindustrylist[k])
    nyindustry.append(new)
    k += 1
    
#4
nasnames = []
l = 0
for thing in nasnamelist:
    new = parse_one(nasnamelist[l])
    nasnames.append(new)
    l += 1
    
#5
nassector = []
m = 0
for thing in nassectorlist:
    new = parse_one(nassectorlist[m])
    nassector.append(new)
    m += 1
    
#6
nasindustry = []
n = 0
for thing in nasindustrylist:
    new = parse_one(nasindustrylist[n])
    nasindustry.append(new)
    n += 1
```
